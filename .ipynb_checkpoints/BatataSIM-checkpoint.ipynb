```julia
using PyPlot
using Distributions
PyPlot.matplotlib[:rc]("text", usetex=true)
```

```julia
a₁₁ = 11.993
a₂₂ = 32.749
a₁₂ = a₁₁ + a₂₂ 

a₁ = a₁₁/a₁₂ 
a₂ = a₂₂/a₁₂

λ₁ = 47.13
λ₂ = 1136.19

println("a₁ = ",a₁)
println("a₂ = ",a₂)
```

```julia
x = linspace(10,400,50)

S(S₀,x)= S₀*((a₁*(exp(-x/λ₁))) +  (a₂*(exp(-x/λ₂))))

ylabel(L"$S(1,x)$",fontsize=20,rotation=360, labelpad=20)
xlabel(L"$x$",fontsize=20)
ylim(0.3,1)

plot(x,S(1,x))
grid("on")
```

## Cheking Poisson function

```julia
k = linspace(0,20,21)

p₁ = Poisson(2)
p₂ = Poisson(5)
p₃ = Poisson(8)


cpois₁ = cdf(p₁,k)
cpois₂ = cdf(p₂,k)
cpois₃ = cdf(p₃,k)


ylim(0,1.3)

plot(k,cpois₁,"r",label=L"$cpois_1$")
plot(k,cpois₂,"b",label=L"$cpois_2$")
plot(k,cpois₃,"g",label=L"$cpois_3$")
legend(loc=4)

xlabel(L"$k$",fontsize=20)
```

```julia
k = linspace(0,20,21)

p₁ = Poisson(2)
p₂ = Poisson(5)
p₃ = Poisson(8)

dpois₁ = pdf(p₁,k)
dpois₂ = pdf(p₂,k)
dpois₃ = pdf(p₃,k)

plot(k,dpois₁,"r",label=L"$dpois_1$")
plot(k,dpois₂,"b",label=L"$dpois_2$")
plot(k,dpois₃,"g",label=L"$dpois_3$")
legend(loc=4)

xlabel(L"$k$",fontsize=20)
```
